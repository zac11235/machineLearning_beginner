梯度下降法
梯度下降法（Gradient Descent）是一种常用的优化算法，用于寻找函数的最小值。它在机器学习和数值优化等领域广泛应用。下面通过一个简单的例子来说明梯度下降法的应用。

假设我们要最小化一个简单的二次函数 f(x) = x^2，其中 x 是变量。我们知道该函数的最小值为 0，现在我们使用梯度下降法来找到这个最小值。

首先，我们需要定义目标函数和其梯度。在这个例子中，目标函数为 f(x) = x^2，梯度为 f'(x) = 2x。

然后，我们选择一个初始值作为起点，例如 x = 3。接下来，根据梯度下降法的更新规则，我们迭代地更新 x 的值，直到达到停止条件（如达到最大迭代次数或梯度足够接近于零）

```python
def gradient_descent(learning_rate, max_iterations):
    # 初始值
    x = 3

    for i in range(max_iterations):
        # 计算梯度
        gradient = 2 * x
        print(f"gradient  {gradient}")
        # 更新x的值
        x = x - learning_rate * gradient
        print(f"x:{x}")
    return x

# 设置学习率和最大迭代次数
learning_rate = 0.1
max_iterations = 100

# 调用梯度下降法函数
result = gradient_descent(learning_rate, max_iterations)

print("最小值点：", result)
print("最小值：", result**2)
```

```python
#求解最大值y=-2**x+x
def gradient_descent(learning_rate, max_iterations):
    # 初始值
    x = -4

    for i in range(max_iterations):
        # 计算梯度
        gradient = -2 * x+1
        print(f"gradient  {gradient}")
        # 更新x的值
        x = x + learning_rate * gradient
        print(f"x:{x}")
    return x

# 设置学习率和最大迭代次数
learning_rate = 0.1
max_iterations = 50

# 调用梯度下降法函数
result = gradient_descent(learning_rate, max_iterations)

print("最大值点：", result)
print("最大值：", -result**2+result)
```

让我们通过一个简单的例子来演示牛顿法的应用。

假设我们要求解函数 f(x) = x^2 - 4 的最小值。我们可以使用牛顿法来找到这个最小值点。

首先，我们需要计算函数的一阶和二阶导数：
f'(x) = 2x
f''(x) = 2

然后，我们选择一个初始点 x₀。我们假设 x₀ = 2。

接下来，我们可以使用牛顿法的迭代公式来更新优化变量 x：
xₖ₊₁ = xₖ - f'(xₖ) / f''(xₖ)

我们将进行多次迭代，直到满足停止准则。在这个例子中，我们可以设置一个迭代次数上限，或者停止条件可以是 f'(x) 的绝对值小于某个预定的容差。

请注意，牛顿法的收敛性和结果质量可能会受到初始点选择的影响。在实际应用中，需要根据具体问题来选择合适的初始点和停止准则，以获得满意的结果。
### 实际中降得非常快，起码在这个阶段是这个样子的

```python
def newton_method(f, f_prime, f_double_prime, x0, max_iter=100, tol=1e-6):
    x = x0
    for i in range(max_iter):
        f_prime_x = f_prime(x)
        print(f_prime_x)
        f_double_prime_x = f_double_prime(x)
        print(f_double_prime_x)
        if abs(f_prime_x) < tol:
            break
        x = x - f_prime_x / f_double_prime_x
        print(x)
    return x

# 定义目标函数及其一阶、二阶导数
def f(x):
    return x**2 -x+4

def f_prime(x):
    return 2 * x-1

def f_double_prime(x):
    return 2

# 设置初始点和迭代次数上限
x0 = 10
max_iter = 100

# 使用牛顿法求解最小值
solution = newton_method(f, f_prime, f_double_prime, x0, max_iter)

print("Solution:", solution)
print("Minimum value:", f(solution))
```

```python
class Solution:
    def runningSum(nums):
        self_sum=[nums[0]]
        n=nums[0]
        for i in range(1,len(nums)):   
            n=n+nums[i]
            self_sum.append(n)
        return self_sum
        """
        :type nums: List[int]
        :rtype: List[int]
        """
num = [1,2,3,4]
a=Solution
result=a.runningSum(nums=num)
print(result)
```

```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.svm import SVC

# 创建一个二分类的数据集
X, y = make_blobs(n_samples=100, centers=2, random_state=0)

# 创建一个SVM分类器
svm = SVC(kernel='linear')

# 训练SVM分类器
svm.fit(X, y)

# 提取支持向量的索引
support_vector_indices = svm.support_

# 提取支持向量的坐标
support_vectors = svm.support_vectors_

# 提取拉格朗日乘子
dual_coefficients = svm.dual_coef_

# 提取偏置项
intercept = svm.intercept_

# 打印结果
print("Support Vector Indices:", support_vector_indices)
print("Support Vectors:\n", support_vectors)
print("Dual Coefficients:", dual_coefficients)
print("Intercept:", intercept)
```
